```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\Lenovo\OneDrive\Desktop\covid_19_india.csv")
df
```

```python
df1=df.drop(['Time','ConfirmedIndianNational', 'ConfirmedForeignNational','Unnamed: 9'], axis=1)
```

```python
active=df.Confirmed-(df.Cured+df.Deaths)
df1['Active']=active
df1.head()
```

```python
df1.sort_values(["Active","Date"], axis=0,ascending=True, inplace=True)
df1
```

```python
print("STATE WITH MAXIMUM NUMBER OF CASES")
p=df1['Active'].max() 
print(p)
print(df1.iloc[-2]['State/UnionTerritory'])
print("\n")
print("STATE WITH MINIMUM NUMBER OF CASES")
q=df1['Active'].min()  
print(q)
print(df1.iloc[0]['State/UnionTerritory'])
```

```python
df1['Ratio'] = df1['Deaths']/df1['Cured'].sum()
df1
```

```python
df1.sort_values(["Ratio","Date"], axis=0,ascending=True, inplace=True)
df1
```

```python
print("STATE WITH MAXIMUM DEATH TO CURE RATIO")
print(df1.iloc[-2]['State/UnionTerritory'])
print("\n")
print("STATE WITH MINIMUM DEATH TO CURE RATIO")
print(df1.iloc[0]['State/UnionTerritory'])
```

```python
df1.rename(columns = {'State/UnionTerritory':'State'}, inplace = True)
df1_list=['Delhi']
df1_sv=df1[df1.State.isin(df1_list)]
df1_sv
```

```python
print("DETAILS OF DELHI")
print("CURED vs DATE")
plt.rcParams["figure.figsize"] = (40,50)
x=df1_sv['Date'].tail(148)
y=df1_sv['Cured'].tail(148)
plt.xlabel('DATE', fontsize=40)
plt.ylabel('CURED', fontsize=100)
plt.xticks(rotation=90)
plt.bar(x,y)
```

```python
print("DEATH vs DATE")
plt.rcParams["figure.figsize"] = (40,50)
x=df1_sv['Date'].tail(148)
y=df1_sv['Deaths'].tail(148)
plt.xlabel('DATE', fontsize=40)
plt.ylabel('DEATHS', fontsize=100)
plt.xticks(rotation=90)
plt.bar(x,y,color='red')
```

```python
print("ACTIVE vs DATE")
plt.rcParams["figure.figsize"] = (40,50)
x=df1_sv['Date'].tail(148)
y=df1_sv['Active'].tail(148)
plt.xlabel('DATE', fontsize=40)
plt.ylabel('ACTIVE', fontsize=100)
plt.xticks(rotation=90)
plt.bar(x,y,color='green')
```

```python
plt.rcParams["figure.figsize"] = (10,5)
p=df1_sv.iloc[-1]['Active']
q=df1_sv.iloc[-1]['Deaths']
r=df1_sv.iloc[-1]['Cured']
task=[p,q,r]
my_labels = 'ACTIVE CASES','DEATHS','CURED'
my_colors = ['lightblue','lightsteelblue','silver']
my_explode = (0.08, 0.03, 0.1)
plt.pie(task, labels=my_labels, autopct='%1.1f%%', startangle=15, shadow = True, colors=my_colors, explode=my_explode)
plt.title('DELHI')
plt.axis('equal')
plt.show()
```
